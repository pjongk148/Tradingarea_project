```python
import pandas  as  pd
```

```python
#데이터  준비 및 컬럼 재지정, 인덱스 재지정
df = pd.read_csv("토지지목별현황_20210527235436.csv", encoding="cp949")
df.columns = list(df[:1].values)
df = df.drop(0).reset_index(drop=True).iloc[:,1:]
```

```python
#행정구역(구동)별(2) 컬럼명 동 으로 변경
df.columns = ['동', '대지', '주차장', '도로', '공원', '체육용지']
df.tail(20)
```

```python
#데이터 사전탐색
#int값으로 변경 요망
df.info()

for i in ['대지', '주차장', '도로', '공원', '체육용지']:
    df[i] = df[i].astype(int)
```

```python
#int확인
df.info()
```

```python
#동  아닌 값  제거
df.head(50)
df = df[-(df["동"]=="소계")].reset_index(drop=True)
```

```python
#반복문으로 매산로->매산동 변경, 팔달로->행궁동
z = []
for i in list(df["동"]):
    if "매산로" in i:
        i = i.replace(i,"매산동")
        z.append(i)
    elif  "팔달로" in i:
        i = i.replace(i,"행궁동")
        z.append(i)
    else:
        z.append(i)
    
df["동"] =z
```

```python
#매산동,  매탄동 병합
df = df.groupby(by ="동").sum()
```

```python
#대지 대비  주차장 면적 비율 파생변수 생성
df["주차장면적/대지(%)"] = (df["주차장"]/df["대지"]*100).round(2)
df
```

```python
#체육용지 유무 바이너리 변수 생성

z = []
for  i in list(df["체육용지"]):
    if i >0:
        z.append(1)
    else:
        z.append(0)
        
df["체육용지유무"]=z
df.head()
```

```python
#전처리 완성본 저장
df.to_csv("전처리완성본/토지지목별 현황 전처리 완성본.csv")
```
