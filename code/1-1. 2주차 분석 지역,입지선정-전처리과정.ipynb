# 지역화폐가맹점 현황 데이터 전처리-동, 업종

## 1.전처리

### 1-1.데이터 결측값  및 컬럼이름 변경

```python
#데이터 준비
data = pd.read_csv("지역화폐가맹점현황.csv", encoding="cp949")
data.head()
```

```python
#업종명(종목명) 업종명으로 변수이름 변경
data.rename(columns={"업종명(종목명)":"업종명"}, inplace=True)

data.head()
```

```python
#데이터 결측값  확인
#업종명 및 소재지지번주소에 결측값 있음
data.info()
```

```python
#결측값 제거 >인덱스 설정>인덱스컬럼제거
df = data.dropna(axis=0).reset_index().iloc[:,2:-1]
df
```

### 1-2.업종명 전처리

```python
#일반휴게음식점만  필터링
dff = df[df["업종명"].str.contains("일반휴게음식")].reset_index(drop=True)
dff
```

```python
#일반휴게음식-*** 형식이 아닌 데이터 필터링
gar = dff[-dff["업종명"].str.contains("일반휴게음식-")].reset_index(drop=True)
gar
```

```python
# 일반휴게음식- 형식이 아닌 값 정규처리
gar_2 = gar["업종명"].replace("일반휴게음식 / 일반한식","일반휴게음식-일반한식").str.replace("휴게음식","휴게음식-")
gar_2 = gar_2.reset_index(drop=True)
```

```python
gar_3 = gar_2.str.replace("일반휴게음식- - ","일반휴게음식-").str.replace("일반휴게음식--","일반휴게음식-")
final = gar_3.str.split("-")
final
gar_3
```

```python
#업종소분류명 분류되는지 확인
b = []
for i in notebook.tqdm(range(len(final))):
    b.append(final[i][1])
```

```python
#gar 업종명 값 처리완료된 자료로 대체
gar["업종명"] = gar_3
gar["업종명"]
gar
```

```python
#정규처리 안해도 되는 좋은 자료 일반휴게음식점 명단
good = dff[dff["업종명"].str.contains("일반휴게음식-")].reset_index(drop=True)
good
```

```python
#정규처리 안해도 되는 좋은 자료+ 정규처리 자료 
df_0 = pd.concat([good,gar]).reset_index(drop=True)
df_0
```

```python
##업종명을 업종대분류 컬럼, 업종소분류 으로 나누기

#"-"기준으로 나눈다
standard = df_0["업종명"].str.split("-")

#업종명
a = []
for i in notebook.tqdm(range(len(df_0))):
    a.append(standard[i][0])
    
#업종소분류명
b = []
for i in notebook.tqdm(range(len(df_0))):
    b.append(standard[i][1])
    
```

```python
#컬럼생성
df_0["업종대분류"] = a
df_0["업종소분류"] = b

#컬럼순서변경 df_0 ->분석용 데이터!
cols = ["상호명","업종명","업종대분류","업종소분류","소재지도로명주소","소재지지번주소","우편번호","위도","경도"]
df_0 = pd.DataFrame(df_0, columns=cols)
df_0
```

### 1-3.소재지지번주소 전처리

```python
#매산로는 모두 매산동, 팔달로는 모두 행궁동, "로" 로 되어있는 소재지지번주소는 팔달로 매산로 만 있다.
#매산로 팔달로 매산동 행궁동으로 전처리 요망
df_0[-df_0["소재지지번주소"].str.contains("매산로|팔달로|동")]


df_0[df_0["소재지지번주소"].str.contains("매산로|팔달로")]
```

```python
#반복문 돌려서 동|팔달로|매산로 가져오기
spl=df_0["소재지지번주소"].str.split(" ")
z=[]
for i in range(len(df_0)):
    z.append(spl[i][3])
```

```python
df_0["동명"] = z
cols = ["상호명","업종명","업종대분류","업종소분류","소재지도로명주소","소재지지번주소","동명","우편번호","위도","경도"]
df_0 = pd.DataFrame(df_0, columns=cols)
df_0
```

```python
#데이터 전처리 완료된 데이터 파일로 저장
df_0.to_csv("sample.csv")
```

```python
df = pd.read_csv("sample.csv")
df["동명"].unique()
```

```python
#1동, 2동, ~~로 다 동으로 통합
df["동명"][df["동명"].str.contains("매산")].unique()
df["동명"] = df["동명"].replace(['매산로1가', '매산로2가', '매산로3가'],"매산동")

#
df["동명"][df["동명"].str.contains("팔달")].unique()
df["동명"] = df["동명"].replace(['팔달로1가', '팔달로3가', '팔달로2가'],"행궁동")

#
df["동명"][df["동명"].str.contains("매탄")].unique()
df["동명"] = df["동명"].replace(['매탄동', '매탄3동', '매탄4동'],"매탄동")

#
df["동명"][df["동명"].str.contains("영통")].unique()
df["동명"] = df["동명"].replace(['영통동', '영통2동'],"영통동")

#
df["동명"][df["동명"].str.contains("화서")].unique()
df["동명"] = df["동명"].replace(['화서동', '화서1동'],"화서동")

#
df["동명"][df["동명"].str.contains("세류")].unique()
df["동명"] = df["동명"].replace(['세류동', '세류2동', '세류3동'],"세류동")
```

```python
#컬럼첫번째  제거
df=df.iloc[:,1:]
```

### 1-4. 전처리 완료된 데이터 파일로 저장(이름=sample로)

```python
#인덱스 제외하고 저장
df.to_csv("전처리 완성본/지역화페 전처리 완성본.csv", index=False)
```
