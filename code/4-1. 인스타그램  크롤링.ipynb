## 워드클라우드

```python
#이중  리스트 반복문으로  탈출 및 공백 제거
one_list = []
for i in crawling:
    for  z  in  i:
        one_list.append(z.replace("[]",""))
        
one_list
```

```python
#단어  카운트
from collections import Counter
count = Counter(one_list)
count
word_dict = dict(count) #딕셔너리 형태로 저장
word_dict
```

```python
#카운트 수가 3보다 작으면 0으로 바꿔라
ze=[]
for z in list(word_dict.values()):
    if z <3:
        z =0
        ze.append(z)
        
    else:
        ze.append(z)
        
        
#처리한 dict의 value값을 새로 넣어줘라
#공백의 딕트에 key value값을 새로 추가한다.
word_dict_modify={}

for i in range(len(word_dict.keys())):
    word_dict_modify[list(word_dict.keys())[i]] = ze[i]
```

```python
#처리값 확인
word_dict_modify
```

```python
#특정  단어 제외
stopword = {"#수원데이트","#수원맛집","#수원이자카야","#영통데이트","#수원일식","#신동맛집","#신동데이트","#영통","#수원","#영통맛집",
           "#매탄동맛집", "#먹스타그램","#선팔하면맞팔","#선팔","#수원맛집추천","#맛집","#수원회식장소","#인계동술집" }
for word in stopword:
    word_dict_modify.pop(word)
```

```python
#워드클라우드
from wordcloud import WordCloud
wordcloud = WordCloud(font_path="C:/Windows/Fonts/malgun.ttf", width=500, height=500, background_color="white").generate_from_frequencies(word_dict_modify)

plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
