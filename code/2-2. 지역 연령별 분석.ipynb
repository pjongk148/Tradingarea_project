# 인사이트

#### + 상권이 좋은 곳은 인계동, 매탄동 , 영통동, 권선동 이었다
#### + 거주인구가 가장  많은 곳은 영통동> 매탄동>권선동>인계동이다.

#### + 하지만 거주인구를 살펴보면 인계동은 적은 것을 알 수 있다. 인계동은 상권형성이 잘 되었지만 거주인구가 적은 곳이다.
#### + 따라서 인계동은 거주인구와 매출이 크게 관계성이 없는것으로 보인다



###  결과

#### + 인계동
    - 20대의 비율이 높다
    - 30대_남   >  20대_여   >  20대_남의 비율이 높다
    - 상대적으로 젊은 연령층 분포가 많아 주점, 칵테일바가 잘 될 것으로 보인다
    
    
#### + 매탄동
    - 20대 와 40대의 비율이 높다
    - 20대_남   >  40대_여의 비율이 높다
    
    
    
#### + 영통동
    - 연령대가 고루 분포되어 있다
    - 20대_남   >  40대_여   >  30대_남  비율이 높다
    
    
#### + 권선동
    - 연령대가 고루 분포되어 있다
    - 3,40대 남녀 비율이 높다
