```python
df = pd.read_excel("역별 유동인구 현황 2020년12월말기준.xls")
df
```

```python
df_1 = df.groupby("지역").sum()
df_1["승강차합계"] = df_1["여객_승차인원"]+df_1["여객_강차인원"]
df_1.sort_values("승강차합계",ascending=False)
```
