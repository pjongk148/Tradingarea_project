```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv("202012_연령별인구현황_연간.csv", encoding="cp949")
df
```

```python
#데이터 사전탐색, 계산을 위해 int값으로 변경 요망
df.info()

#그전에 쉼표 빼줘야 타입 변경가능
cols = df.columns[1:]
list(cols)

for i in list(cols):
    df[i]=df[i].str.replace(",","")
    

#int로변경
for i in list(cols):
    df[i]=df[i].astype(int)
```

```python
#변경 확인
df.info()
```

```python
#컬럼 깔끔하게 변경
df.columns = df.columns.str.replace("2020년_","").str.replace("거주자_","").str.replace("20~29세","20대").str.replace("30~39세","30대").str.replace("40~49세","40대")
df.head()
```

```python
#행정구역 번호및 괄호 제거, 시, 구 명 제거 (동만 남기기)
df["행정구역"] = df["행정구역"].str.replace("\W|\d", "")
df["행정구역"] = df["행정구역"].str.replace("[ㄱ-ㅣ,가-힣]+시|[ㄱ-ㅣ,가-힣]+구","")
df
```

```python
#전처리 완성본
df_final = df[-(df["행정구역"]=="")].groupby(by="행정구역").sum().reset_index()
df_final
```

```python
#완성본 파일로 저장
df_final.to_csv("전처리 완성본/지역연령별 데이터 전처리 완성본.csv", index=False)
```
